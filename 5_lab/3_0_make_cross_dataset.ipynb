Дальше будут попытки finenuting'а нескольких моделей, чтобы их как-то можно было бы сравнить между собой, сразу разделим текст на train и test

```python
import pandas as pd
from datasets import Dataset
```

```python
df = pd.read_parquet('./__input__/mathced.parquet')

df
```

```python
df[df['text'] == df['target_text']]
```

```python
df = df[df['text'] != df['target_text']]
```

```python
ds = Dataset.from_pandas(df)
ds = ds.train_test_split(test_size=0.1) # type: ignore

ds
```

```python
ds['train'].to_parquet('./__output__/matched_train.parquet') # type: ignore 
ds['test'].to_parquet('./__output__/matched_test.parquet') # type: ignore 
```
